# Get total_passenger

```python
df2 = df.copy()
```

```python
dayWeek = df2[df2[['day', 'week']].duplicated() == False][['day', 'week']]
```

```python
df2[(df2['day'] == 'Mon') & (df2['week'] == 0)]['id'].count()
```

```python
for _, dw in dayWeek.iterrows():
    N_the_day = df2[(df2['day'] == dw.day) & (df2['week'] == dw.week)]['id'].count()
    indexes_the_day = df2[(df2['day'] == dw.day) & (df2['week'] == dw.week)].index
    for i in indexes_the_day:
        df2.at[i, 'total_passenger'] = N_the_day
```

```python
df2.to_csv('cleaned_data.csv', index=False)
```
